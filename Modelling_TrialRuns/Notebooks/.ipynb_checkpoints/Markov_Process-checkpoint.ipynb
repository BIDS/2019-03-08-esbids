## Implementation of the Differential Equations version of the Continuous Markov Chain 
This creates 9 states 

```julia
#1. The elements on the main diagonal are all strictly negative.
#2. The elements off the main diagonal are non-negative.
#3. Each row sums to zero.

#=
G= [-3 3 0;
1 -1 0;
0  2 -2]
=#
stay1 = 2 # prob -> 2 = 3/4, prob -> 3 = 1/4
stay2 = 3 # prob -> 1 = 1/2, prob -> 3 = 1/2
stay3 = 5 # prob -> 2 = 2/3, prob -> 1 = 1/3

G= [-1/stay1 (3/4*1/stay1) (1/4*1/stay1); 
    (1/2*1/stay2) -1/stay2 (1/2*1/stay2); 
    (1/3*1/stay3)  (2/3*1/stay3) -1/stay3]

u0 = Matrix{Float64}(I, 3, 3)
tspan = (0.0,50)
f(u,p,t) = u*G # where u = P(t) from the readings and G (generator matrix) is also called Q 
prob = ODEProblem(f,u0,tspan)
```

```julia
sol = solve(prob)
;
```

```julia
#plot(sol)
p = plot()
for n in 1:9
    plot([r[n] for r in sol.u])
end
```

## Implementation of the closed form solution. 
Takes the same timestamps that DIfferentialEquations used to solve and directly evaluates the closed form solution

```julia
time_vector = sol.t
;
```

```julia
result = Vector{Matrix{Float64}}(undef, length(time_vector))
for (ix, t) in enumerate(time_vector)
    result[ix] = u0*exp(G*t)
end
```

```julia
result
;
```

```julia
p = plot()
for n in 1:9
    plot([r[n] for r in result])
end

#=
p = plot()
for n in 1:9
    plot!([r[n] for r in result])
end
display(p)
=#
```
